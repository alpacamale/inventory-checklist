```python
import pandas as pd

input_file = "data/data.csv"
encoded_file = "data/encoded.csv"
output_file = "data/items.csv"

with open(input_file, "r", encoding="euc-kr") as f:
    text = f.read()

with open(encoded_file, "w", encoding="utf-8") as f:
    f.write(text)

df = pd.read_csv(encoded_file)
df.head()
```

```python
df = df[["품목코드", "품목명", "단위"]]
df.head()
```

```python
df = df.dropna()
df.head()
```

```python
df["품목코드"] = df["품목코드"].astype(int)
df.head()
```

```python
df.to_csv(output_file, index=False)
```
